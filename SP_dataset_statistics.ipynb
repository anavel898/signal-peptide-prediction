```python
SP_bench_lengths = []

for sp in SP_bench["SP cleavage-site annotation"]:
  i = 0
  spLength = 0
  while sp[i] != 'N':
    spLength+=1
    i = i+1
  SP_bench_lengths.append(spLength)

SP_bench_lengths_series = pd.Series(SP_bench_lengths)
SP_bench_lengths_series
```

# Extracting sequence for creating the sequence logo

```python
from google.colab import files
```

```python
# 15-residue long motif [-13,+2] around the cleavage site is extracted
def getSeqForLogo(SP_df, SP_len_list, file_name):
  sequencesList = []
  cut = " "
  for seq, length in zip(SP_df["Sequence (first 50 N-terminal residues)"], SP_len_list):
    cut = seq[0:(length+2)]
    cut = cut[-15:]
    sequencesList.append(cut)
  with open(file_name, 'w') as f:
    i=1
    for seq in sequencesList:
      seq = ">Seq"+str(i)+"\n"+seq+"\n"
      f.write(seq)
      i+=1
    files.download(file_name)
  return
```

# Comparative aa composition of SPs: benchmark vs. training vs. swissprot

```python
import re
```

```python
#contents of the uploaded file are obtained from statistics provided here: https://web.expasy.org/docs/relnotes/relstat.html
!wget https://raw.githubusercontent.com/anavel898/signal-peptide-prediction/main/datasets/aa-composition-uniprot.txt
```

Loaded file is of following structure:

Leu (L) 9.65

Ala (A) 8.25

Gly (G) 7.07

Val (V) 6.86

 ....

So, I'm using regex to get the dictionary where key is the amino acid symbol and the value is it's frequency in the swissprot database.

```python
regex = "\([A-Z]\)"
percentage_regex = "[0-9]*\.[0-9]*"
aa_list = []
aa_comp_uniprot = {}
with open('aa-composition-uniprot.txt', 'r') as aa_comp:
  for line in aa_comp:
    x = re.search(regex, line)
    y = re.search(percentage_regex,line)
    aa_list.append(x.group()[1])
    aa_comp_uniprot[x.group()[1]] = float(y.group())
print(aa_comp_uniprot)
```

```python
#creating a dictionary of aa frequencies in SPs of the training dataset
tr_aa_count = {}
all_residues = 0
for seq, length in zip(SP_training["Sequence (first 50 N-terminal residues)"], SP_lengths):
  SPs = seq[0:length]
  for residue in SPs:
    if residue not in tr_aa_count:
      tr_aa_count[residue]=1
    else:
      tr_aa_count[residue]+=1
    all_residues+=1
tr_aa_freq = {}
for aa in aa_list:
  tr_aa_freq[aa] = tr_aa_count[aa]/all_residues*100
print(tr_aa_freq)
```

```python
#creating a dictionary of aa frequencies in SPs of the benchmarking dataset
bench_aa_count = {}
all_residues = 0
for seq, length in zip(SP_bench["Sequence (first 50 N-terminal residues)"], SP_bench_lengths):
  SPs = seq[0:length]
  for residue in SPs:
    if residue not in bench_aa_count:
      bench_aa_count[residue]=1
    else:
      bench_aa_count[residue]+=1
    all_residues+=1
bench_aa_freq = {}
for aa in aa_list:
  bench_aa_freq[aa] = bench_aa_count[aa]/all_residues*100
print(bench_aa_freq)
```

```python
aa_composition = pd.DataFrame.from_dict(aa_comp_uniprot, orient='index', columns = ['UniProt'])
aa_composition
```

```python
tr_df = pd.DataFrame.from_dict(tr_aa_freq, orient='index', columns = ["Training"])
tr_df
```

```python
bench_df = pd.DataFrame.from_dict(bench_aa_freq, orient='index', columns = ["Benchmark"])
bench_df
```

```python
aa_composition["Training"] = tr_df["Training"]
aa_composition["Benchmark"] = bench_df["Benchmark"]
aa_composition["Residues"] = aa_list
aa_composition
```

```python
melted_aa_comp = aa_composition.melt(id_vars = 'Residues', var_name = "Dataset", value_name="Frequencies(%)")
melted_aa_comp
```

```python
sns.barplot(melted_aa_comp, x=melted_aa_comp['Residues'], y = melted_aa_comp['Frequencies(%)'], hue = melted_aa_comp['Dataset'])
plt.legend(title='Residue frequencies', fontsize = 'large', title_fontsize = "10")
sns.set(rc={'figure.figsize':(16,6)})
plt.show()
```
